```python
import mlflow
from synapse.ml.predict import MLFlowTransformer
from pyspark.sql.functions import col, when

ws = "HospAdmission"
lh = "Gold_LH"
table = "hosp_training_data"
    

df = spark.read.format("delta").load(
     f"abfss://{ws}@onelake.dfs.fabric.microsoft.com/{lh}.Lakehouse/Tables/{table}"# Your input table filepath here
    )

model = MLFlowTransformer(
    inputCols=["Age","Gender","ChronicCondition","Department","FundingType","LengthOfStayDays","GlucoseLevel","Cholesterol","SystolicBP","DiastolicBP","BMI","PastAdmissions6M","AvgPastLengthOfStay"], # Your input columns here
    outputCol="predictions", # Your new column name here
    modelName="reAdmissionRisk-model", # Your model name here
    modelVersion=1 # Your model version here
)
df = model.transform(df)

df = df.withColumn('AdmissionRisk', when(col('predictions') == 0 ,'Low Risk').otherwise('High Risk'))
df = df.drop('predictions')

df.write.format('delta').mode("overwrite").save(
    f"abfss://{ws}@onelake.dfs.fabric.microsoft.com/{lh}.Lakehouse/Tables/predicted_table" # Your output table filepath here
)
```
